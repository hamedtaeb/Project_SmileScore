```python
# Define Target and Features

TARGET = "happiness_score"
FEATURES = [
    "gdp_per_capita",
    "social_support",
    "healthy_life_expectancy",
    "freedom_to_make_life_choices",
    "generosity",
    "perceptions_of_corruption",
]
```

```python
# Optional safety: ensure all features exist and are numeric
for col in FEATURES + [TARGET]:
    if col not in df.columns:
        raise ValueError(f"Column missing in df: {col}")
    df[col] = pd.to_numeric(df[col], errors="coerce")
df = df.dropna(subset=FEATURES + [TARGET, "year"])
```

```python
#Train/Test split by year

latest_year = int(df["year"].max())
train = df[df["year"] < latest_year].copy()
test  = df[df["year"] == latest_year].copy()

X_train, y_train = train[FEATURES], train[TARGET]
X_test,  y_test  = test[FEATURES],  test[TARGET]

print(f"Train years: {sorted(train['year'].unique())}")
print(f"Test year:   {sorted(test['year'].unique())}")
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
```

```python
# Helpers
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def summarize(model_name, y_true, y_pred):
    return {"model": model_name,
            "R2": r2_score(y_true, y_pred),
            "RMSE": rmse(y_true, y_pred)}
```

```python
# Linear Regression (with scaling)
lr_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LinearRegression())
])
lr_pipe.fit(X_train, y_train)
pred_test_lr = lr_pipe.predict(X_test)
```

```python
# Random Forest (baseline + light tuning via CV on TRAIN only)
rf_base = RandomForestRegressor(
    n_estimators=400,
    random_state=42,
    n_jobs=-1
)
rf_base.fit(X_train, y_train)
pred_test_rf_base = rf_base.predict(X_test)
```

```python
# Light hyperparameter search
param_distributions = {
    "n_estimators": [300, 400, 600, 800],
    "max_depth": [None, 6, 8, 10, 12, 16],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None],
}
cv = KFold(n_splits=3, shuffle=True, random_state=42)

rf = RandomForestRegressor(random_state=42, n_jobs=-1)
rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_distributions,
    n_iter=25,
    scoring="neg_root_mean_squared_error",
    cv=cv,
    n_jobs=-1,
    random_state=42,
    verbose=0
)
rf_search.fit(X_train, y_train)
rf_best = rf_search.best_estimator_
pred_test_rf_best = rf_best.predict(X_test)

print("Best RF params:", rf_search.best_params_)
```

```python
# Compare on TEST (latest year only)
rows = []
rows.append(summarize("LinearRegression (Test)", y_test, pred_test_lr))
rows.append(summarize("RandomForest Base (Test)", y_test, pred_test_rf_base))
rows.append(summarize("RandomForest Tuned (Test)", y_test, pred_test_rf_best))

metrics_df = pd.DataFrame(rows)
display(metrics_df)

# Pick best by TEST RMSE (since we have only Train/Test per mentor guidance)
test_rmse = {
    "LR": rmse(y_test, pred_test_lr),
    "RF_BASE": rmse(y_test, pred_test_rf_base),
    "RF_TUNED": rmse(y_test, pred_test_rf_best),
}
best_key = min(test_rmse, key=test_rmse.get)
if best_key == "LR":
    best_model, y_pred_test = lr_pipe, pred_test_lr
elif best_key == "RF_BASE":
    best_model, y_pred_test = rf_base, pred_test_rf_base
else:
    best_model, y_pred_test = rf_best, pred_test_rf_best

print(f"Best on TEST by RMSE: {best_key}")
print(f"Test R:  {r2_score(y_test, y_pred_test):.4f}")
print(f"Test RMSE:{rmse(y_test, y_pred_test):.4f}")
```

```python
# Diagnostics: Actual vs Predicted (Test)
# -------------------------------
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_test, alpha=0.6)
lims = [min(y_test.min(), y_pred_test.min()), max(y_test.max(), y_pred_test.max())]
plt.plot(lims, lims, "--", linewidth=1)
plt.xlim(lims); plt.ylim(lims)
plt.xlabel("Actual Happiness Score")
plt.ylabel("Predicted Happiness Score")
plt.title(f"Actual vs Predicted (Test {latest_year})")
plt.tight_layout()
plt.show()
```

```python
# Interpretability 

if isinstance(best_model, RandomForestRegressor):
    importances = pd.Series(best_model.feature_importances_, index=FEATURES).sort_values()
    plt.figure(figsize=(6,4))
    plt.barh(importances.index, importances.values)
    plt.xlabel("Importance")
    plt.title("Random Forest Feature Importance")
    plt.tight_layout()
    plt.show()
```
